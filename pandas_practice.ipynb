```
#Pandas Series:array like object conatining series of values & associated array of data labels called index
import pandas as pd
#from pandas import Series,DataFrame
obj = pd.Series([4,7,-5,3])
print(obj)
print(obj.values)
print(obj.index)
obj2 = pd.Series([4,7,-5,3], index = ['d','b','a','c'])
print(obj2)
print(obj2.index)
print(obj2['a'])
obj2['d']= 6
print(obj2)
print(obj2[obj2>0])
print(obj2*2)
print('b' in obj2)
```

```
#making series out of dictionary
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
print(obj3)
#giving your own keys to series made of dictionary
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata,index = states)
print(obj4)
print(pd.isnull(obj4))
print(pd.notnull(obj4))
```
